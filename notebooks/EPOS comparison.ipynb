```python
%matplotlib inline  
import matplotlib as mpl
import matplotlib.pyplot as plt
from statistics import mean, median, variance 
plt.rcParams['figure.figsize'] = [10, 5]
import pprint

def get_data_from_file(file_name, which):
    data = []
    with open(file_name, 'r') as results:
        for line in results:
            line = line.split(",")
            trip_duration = float(line[4])
            overhead = float(line[6])
            agent = line[7]
            preference = line[8].replace('\r', '').replace('\n', '')
            
            data.append(overhead)
    return data

def get_baseline_data(name):
    return get_data_from_file("results/baselines/overhead_baseline_" + name + ".csv", name)

def get_epos_data(name):
    return get_data_from_file("results/epos/overhead_epos_" + name + ".csv", name)

baseline_min_length = sorted(get_baseline_data("min_length"))
baseline_max_speed = sorted(get_baseline_data("max_speed"))
baseline_balanced = sorted(get_baseline_data("balanced"))
baseline_mixed = sorted(get_baseline_data("mixed"))
epos_altruistic = sorted(get_epos_data("altruistic"))
# epos_selfish = sorted(get_epos_data("selfish"))
epos_1 = sorted(get_epos_data("0_0.9"))

print(len(baseline_min_length))
print(len(baseline_max_speed))
print(len(baseline_balanced))
print(len(baseline_mixed))
# print(len(epos_altruistic))
# print(len(epos_selfish))

plt.plot(baseline_min_length, label="baseline_min_length")
plt.plot(baseline_max_speed, label="baseline_max_speed")
plt.plot(baseline_balanced, label="baseline_balanced")
plt.plot(baseline_mixed, label="baseline_mixed")
plt.plot(epos_altruistic, label="epos_altruistic")
# plt.plot(epos_selfish, label="epos_selfish")
plt.plot(epos_1, label="epos_0_0.9")

plt.legend(loc='best')
plt.show()

print("Average overhead for 'baseline_min_length' is % s" % (mean(baseline_min_length))) 
print("Average overhead for 'baseline_max_speed' is % s" % (mean(baseline_max_speed))) 
print("Average overhead for 'baseline_balanced' is % s" % (mean(baseline_balanced))) 
print("Average overhead for 'baseline_mixed' is % s" % (mean(baseline_mixed))) 
print("Average overhead for 'epos_altruistic' is % s" % (mean(epos_altruistic))) 
# print("Average overhead for 'epos_selfish' is % s" % (mean(epos_selfish))) 
print("Average overhead for 'epos_0_0.9' is % s" % (mean(epos_1))) 
print("******")
print("Median overhead for 'baseline_min_length' is % s" % (median(baseline_min_length))) 
print("Median overhead for 'baseline_max_speed' is % s" % (median(baseline_max_speed))) 
print("Median overhead for 'baseline_balanced' is % s" % (median(baseline_balanced))) 
print("Median overhead for 'baseline_mixed' is % s" % (median(baseline_mixed))) 
print("Median overhead for 'epos_altruistic' is % s" % (median(epos_altruistic))) 
# print("Average overhead for 'epos_selfish' is % s" % (median(epos_selfish))) 
print("******")
print("Variance of overhead for 'baseline_min_length' is % s" % (variance(baseline_min_length))) 
print("Variance of overhead for 'baseline_max_speed' is % s" % (variance(baseline_max_speed))) 
print("Variance of overhead for 'baseline_balanced' is % s" % (variance(baseline_balanced))) 
print("Variance of overhead for 'baseline_mixed' is % s" % (variance(baseline_mixed))) 
print("Variance of overhead for 'epos_altruistic' is % s" % (variance(epos_altruistic))) 
# print("Average overhead for 'epos_selfish' is % s" % (variance(epos_selfish))) 
```
